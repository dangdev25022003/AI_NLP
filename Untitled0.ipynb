```python
bert = AutoModel.from_pretrained('vinai/phobert-base-v2')
```

```python
tokenizer = AutoTokenizer.from_pretrained('vinai/phobert-base-v2')
```

```python
# tokenize and encode sequences in the training set
tokens_train1 = tokenizer.batch_encode_plus(
    train_text1.tolist(),
    max_length = 25,
    pad_to_max_length=True,
    truncation=True
)

# tokenize and encode sequences in the validation set
tokens_val1 = tokenizer.batch_encode_plus(
    val_text1.tolist(),
    max_length = 25,
    pad_to_max_length=True,
    truncation=True
)

# tokenize and encode sequences in the test set
tokens_test1 = tokenizer.batch_encode_plus(
    test_text1.tolist(),
    max_length = 25,
    pad_to_max_length=True,
    truncation=True
)
# tokenize and encode sequences in the training set
tokens_train2 = tokenizer.batch_encode_plus(
    train_text2.tolist(),
    max_length = 25,
    pad_to_max_length=True,
    truncation=True
)

# tokenize and encode sequences in the validation set
tokens_val2 = tokenizer.batch_encode_plus(
    val_text2.tolist(),
    max_length = 25,
    pad_to_max_length=True,
    truncation=True
)

# tokenize and encode sequences in the test set
tokens_test2 = tokenizer.batch_encode_plus(
    test_text2.tolist(),
    max_length = 25,
    pad_to_max_length=True,
    truncation=True
)
```

```python
train_seq1 = torch.tensor(tokens_train1['input_ids'])
train_mask1 = torch.tensor(tokens_train1['attention_mask'])
train_y1 = torch.tensor(train_labels1.tolist())

val_seq1 = torch.tensor(tokens_val1['input_ids'])
val_mask1 = torch.tensor(tokens_val1['attention_mask'])
val_y1 = torch.tensor(val_labels1.tolist())

test_seq1 = torch.tensor(tokens_test1['input_ids'])
test_mask1 = torch.tensor(tokens_test1['attention_mask'])
test_y1 = torch.tensor(test_labels1.tolist())
```

```python
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler

#define a batch size
batch_size = 32

# wrap tensors
train_data1 = TensorDataset(train_seq1, train_mask1, train_y1)

# sampler for sampling the data during training
train_sampler1 = RandomSampler(train_data1)

# dataLoader for train set
train_dataloader1 = DataLoader(train_data1, sampler=train_sampler1, batch_size=batch_size)

# wrap tensors
val_data1 = TensorDataset(val_seq1, val_mask1, val_y1)

# sampler for sampling the data during training
val_sampler1 = SequentialSampler(val_data1)

# dataLoader for validation set
val_dataloader1 = DataLoader(val_data1, sampler = val_sampler1, batch_size=batch_size)
```

```python
# freeze all the parameters
for param in bert.parameters():
    param.requires_grad = False
```

```python
class BERT_Arch(nn.Module):

    def __init__(self, bert):
        super(BERT_Arch, self).__init__()

        self.bert = bert

        # dropout layer
        self.dropout = nn.Dropout(0.1)

        # relu activation function
        self.relu =  nn.ReLU()

        # dense layer 1
        self.fc1 = nn.Linear(768,512)

        # dense layer 2 (Output layer)
        self.fc2 = nn.Linear(512,2)

        #softmax activation function
        self.softmax = nn.LogSoftmax(dim=1)

    #define the forward pass
    def forward(self, sent_id, mask):

        #pass the inputs to the model
        _, cls_hs = self.bert(sent_id, attention_mask=mask, return_dict=False)

        x = self.fc1(cls_hs)

        x = self.relu(x)

        x = self.dropout(x)

        # output layer
        x = self.fc2(x)

        # apply softmax activation
        x = self.softmax(x)

        return x
```

```python
# pass the pre-trained BERT to our define architecture
model = BERT_Arch(bert)

# push the model to GPU
model = model.to(device)
```

```python
# optimizer from hugging face transformers
from transformers import AdamW

# define the optimizer
optimizer = AdamW(model.parameters(),lr = 1e-5)
```

```python
from sklearn.utils.class_weight import compute_class_weight

#compute the class weights
# class_weights = compute_class_weight('balanced', np.unique(train_labels1), train_labels1)
class_weights1 = compute_class_weight(
                                        class_weight = "balanced",
                                        classes = np.unique(train_labels1),
                                        y = train_labels1
                                    )
# class_weights1 = dict(zip(np.unique(train_labels1), class_weights1))
print("Class Weights:",class_weights1)
```

```python
from torch.optim import lr_scheduler
```

```python
# converting list of class weights to a tensor
weights1= torch.tensor(class_weights1*batch_size,dtype=torch.float)

# push to GPU
weights1 = weights1.to(device)

# define the loss function
cross_entropy1  = nn.NLLLoss()

# number of training epochs
epochs = 10
lr_sch = lr_scheduler.StepLR(optimizer, step_size=100, gamma=0.1)
```

```python
# function to train the model
def train():

    model.train()
    total_loss, total_accuracy = 0, 0

    # empty list to save model predictions
    total_preds=[]

    # iterate over batches
    for step,batch in enumerate(train_dataloader1):

        # progress update after every 50 batches.
        if step % 50 == 0 and not step == 0:
            print('  Batch {:>5,}  of  {:>5,}.'.format(step, len(train_dataloader1)))

        # push the batch to gpu
        batch = [r.to(device) for r in batch]

        sent_id, mask, labels = batch

        # clear previously calculated gradients
        model.zero_grad()

        # get model predictions for the current batch
        preds = model(sent_id, mask)

        # compute the loss between actual and predicted values
        loss = cross_entropy1(preds, labels)

        # add on to the total loss
        total_loss = total_loss + loss.item()

        # backward pass to calculate the gradients
        loss.backward()

        # clip the the gradients to 1.0. It helps in preventing the exploding gradient problem
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

        # update parameters
        optimizer.step()

        # model predictions are stored on GPU. So, push it to CPU
        preds=preds.detach().cpu().numpy()

    # append the model predictions
        total_preds.append(preds)

    # compute the training loss of the epoch
    avg_loss = total_loss / len(train_dataloader1)

      # predictions are in the form of (no. of batches, size of batch, no. of classes).
      # reshape the predictions in form of (number of samples, no. of classes)
    total_preds  = np.concatenate(total_preds, axis=0)

    #returns the loss and predictions
    return avg_loss, total_preds
```

```python
import babel
from datetime import datetime
from babel.dates import format_time
import time
```

```python
# function for evaluating the model
def evaluate():

    print("\nEvaluating...")

    # deactivate dropout layers
    model.eval()

    total_loss, total_accuracy = 0, 0

    # empty list to save the model predictions
    total_preds = []

    # iterate over batches
    for step,batch in enumerate(val_dataloader1):

        # Progress update every 50 batches.
        if step % 50 == 0 and not step == 0:

            # Calculate elapsed time in minutes.
            elapsed = format_time(time.time() )

            # Report progress.
            print('  Batch {:>5,}  of  {:>5,}.'.format(step, len(val_dataloader1)))

        # push the batch to gpu
        batch = [t.to(device) for t in batch]

        sent_id, mask, labels = batch

        # deactivate autograd
        with torch.no_grad():

            # model predictions
            preds = model(sent_id, mask)

            # compute the validation loss between actual and predicted values
            loss = cross_entropy1(preds,labels)

            total_loss = total_loss + loss.item()

            preds = preds.detach().cpu().numpy()

            total_preds.append(preds)

    # compute the validation loss of the epoch
    avg_loss = total_loss / len(val_dataloader1)

    # reshape the predictions in form of (number of samples, no. of classes)
    total_preds  = np.concatenate(total_preds, axis=0)

    return avg_loss, total_preds
```

```python
# set initial loss to infinite
best_valid_loss = float('inf')

# empty lists to store training and validation loss of each epoch
train_losses=[]
valid_losses=[]

#for each epoch
for epoch in range(epochs):

    print('\n Epoch {:} / {:}'.format(epoch + 1, epochs))

    #train model
    train_loss1, _ = train()

    #evaluate model
    valid_loss1, _= evaluate()

    #save the best model
    if valid_loss1 < best_valid_loss:
        best_valid_loss = valid_loss1
        torch.save(model.state_dict(), 'saved_weights.pt')

    # append training and validation loss
    train_losses.append(train_loss1)
    valid_losses.append(valid_loss1)

    # it can make your experiment reproducible, similar to set  random seed to all options where there needs a random seed.
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

    print(f'\nTraining Loss: {train_loss1:.3f}')
    print(f'Validation Loss: {valid_loss1:.3f}')
```
